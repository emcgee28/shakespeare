```python
import nltk
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
text = "HAMLET To be, or not to be: that is the question: Whether 'tis nobler in the mind to suffer The slings and arrows of outrageous fortune, Or to take arms against a sea of troubles, And by opposing end them? To die: to sleep;"

blob = TextBlob(text)
result = blob.tags
print(result)
```

```python
result = blob.word_counts
print(result)
```

```python
sorted_items = sorted(result.items(), key=lambda item: item[1], reverse=True)
words, counts = zip(*sorted_items)

plt.figure(figsize=(12, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency in Hamlet Excerpt')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
result = blob.sentiment
print(result)
```
