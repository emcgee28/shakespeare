```python
#plot romeo's word counts
import os
import re
import matplotlib.pyplot as plt
from textblob import TextBlob

directory = r'C:\Users\erinf\OneDrive\Desktop\python\Shakespeare\Romeo'

scene_word_counts = []

for filename in os.listdir(directory):
    match = re.match(r'romeo_act(\d+)scene(\d+)\.txt$', filename.lower())
    if match:
        act = int(match.group(1))
        scene = int(match.group(2))

        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

            blob = TextBlob(content)
            word_count = len(blob.words)

        scene_word_counts.append(((act, scene), word_count))
    else:
        print(f"  Skipped: {filename} (did not match)")

if scene_word_counts:
    scene_word_counts.sort()
    labels = [f"Act {act} Scene {scene}" for (act, scene), _ in scene_word_counts]
    counts = [count for _, count in scene_word_counts]

    plt.figure(figsize=(12, 6))
    plt.bar(labels, counts, color='navy')
    plt.xlabel('Scene')
    plt.ylabel('Number of Words Spoken by Romeo')
    plt.title("Romeo's Word Count by Scene")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig("romeo_word_count.png")  # Optional: saves the plot
    plt.show()
else:
    print("No scenes processed. Check file names and folder path.")
```

```python
import os
import re
import matplotlib.pyplot as plt
from textblob import TextBlob
from collections import defaultdict

# Directories
romeo_dir = r'C:\Users\erinf\OneDrive\Desktop\python\Shakespeare\Romeo'
juliet_dir = r'C:\Users\erinf\OneDrive\Desktop\python\Shakespeare\Juliet'

# Dictionary to store word counts: {(act, scene): {'Romeo': x, 'Juliet': y}}
scene_data = defaultdict(lambda: {'Romeo': 0, 'Juliet': 0})

# Helper function to read and count words
def process_files(directory, character):
    for filename in os.listdir(directory):
        pattern = fr'{character.lower()}_act(\d+)scene(\d+)\.txt$'
        match = re.match(pattern, filename.lower())
        if match:
            act = int(match.group(1))
            scene = int(match.group(2))
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                blob = TextBlob(content)
                word_count = len(blob.words)
            scene_data[(act, scene)][character] = word_count

# Process both sets of files
process_files(romeo_dir, 'Romeo')
process_files(juliet_dir, 'Juliet')

# Sort scenes
sorted_scenes = sorted(scene_data.keys())

# Build data for plotting
labels = [f"Act {act} Scene {scene}" for (act, scene) in sorted_scenes]
romeo_counts = [scene_data[(act, scene)]['Romeo'] for (act, scene) in sorted_scenes]
juliet_counts = [scene_data[(act, scene)]['Juliet'] for (act, scene) in sorted_scenes]

# Plotting
x = range(len(labels))
width = 0.4

plt.figure(figsize=(14, 6))
plt.bar([i - width/2 for i in x], romeo_counts, width=width, label='Romeo', color='skyblue')
plt.bar([i + width/2 for i in x], juliet_counts, width=width, label='Juliet', color='lightcoral')

plt.xlabel('Scene')
plt.ylabel('Word Count')
plt.title("Word Counts by Scene: Romeo vs Juliet")
plt.xticks(ticks=x, labels=labels, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig("romeo_vs_juliet_word_counts.png")
plt.show()
```
